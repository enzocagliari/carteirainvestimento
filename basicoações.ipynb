```python
import pandas as pd
```

```python
import numpy as ny
import matplotlib.pyplot as plt
import yfinance as yf
import seaborn as sns

```

```python
acoes = ["PETR4.SA","PRIO3.SA","SMTO3.SA", "JALL3.SA"]
df = yf.download(acoes, start= "2019-01-01")["Adj Close"]
df
```

```python
df.dropna(inplace= True)
```

```python
df
```

```python
df.plot()
plt.grid()
```

```python
retorno = df.pct_change()
retorno
```

```python
retorno.plot()
```

```python
retorno
```

```python
retorno_acumulado = (retorno + 1).cumprod()
retorno_acumulado
```

```python
retorno_acumulado.plot()
```

```python
carteira = retorno_acumulado[acoes[0]]* retorno_acumulado[acoes[1]]*retorno_acumulado[acoes[2]]* retorno_acumulado[acoes[3]]
carteira
```

```python
carteira.plot()
```
